We will now import the necessary packages to run the project.

```python
import google.generativeai as genai
import pathlib
import textwrap
from IPython.display import display
from IPython.display import Markdown
```

```python
def to_markdown(text):
  text = text.replace('•','*')

  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

input_text = "This is a • sample text with bullet points"
result = to_markdown(input_text)

display(result)
```

```python
from google.colab import userdata
```

```python
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
```

```python
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
for model in genai.list_models():
    print(model)
```

```python
for model in genai.list_models():
  if 'generateContent' in model.supported_generation_methods:
    print(model.name)
```

```python
model = genai.GenerativeModel('gemini-2.0-flash-thinking-exp-01-21')
```

```python
model
```

```python
%%time
response = model.generate_content("Can you tell me the meaning of life?")
```

```python
response.text
```

```python
to_markdown(response.text)
```

Now lets check the order in which the model as generated the output.

```python
%%time

response = model.generate_content("Can you tell me the meaning of life?", stream=True)

for chunk in response:
  print(chunk.text)
  print("_"*100)
```
